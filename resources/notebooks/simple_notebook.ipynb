```python

dbutils.widgets.text("source_schema", "nyc_taxi", "Source Schema")
dbutils.widgets.text("source_catalog", "samples", "Source Catalog")
dbutils.widgets.text("source_table", "trips", "Source Table")
dbutils.widgets.text("target_schema", "krish_dabs_learning", "Target Schema")
dbutils.widgets.text("target_catalog", "main", "Target Catalog")
dbutils.widgets.text("target_table", "destination_data", "Target Table")
dbutils.widgets.text("distance_cutoff", "10", "Cutoff for Distance")

```

```python

source_schema = dbutils.widgets.get("source_schema")
source_catalog = dbutils.widgets.get("source_catalog")
source_table = dbutils.widgets.get("source_table")
target_schema = dbutils.widgets.get("target_schema")
target_catalog = dbutils.widgets.get("target_catalog")
target_table = dbutils.widgets.get("target_table")
distance_cutoff = dbutils.widgets.get("distance_cutoff")

source_full_table = f"{source_catalog}.{source_schema}.{source_table}"
target_full_table = f"{target_catalog}.{target_schema}.{target_table}"
cutoff = float(distance_cutoff)

print(f"Source: {source_full_table}, \nTarget: {target_full_table}, \nCutoff: {cutoff}")
```

```python
df = spark.read.table(source_full_table)

df.show()
```

```python
df = df.filter("trip_distance > 0")
df.show()
```

```python
df = df.groupBy("dropoff_zip").agg({"trip_distance": "avg",
                                    "fare_amount": "avg",
                                    })
df.show()
```

```python
df.write.mode("overwrite").saveAsTable(target_full_table)
```
